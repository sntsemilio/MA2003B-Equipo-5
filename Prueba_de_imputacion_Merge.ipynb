```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# ==================================================================================
# COMPARACI√ìN POR ID: INTERPOLACI√ìN vs ARIMA vs SARIMA vs HOLT-WINTERS
# ==================================================================================

# Variables a imputar
variables_imputar = ['CO (ppm)', 'NO (ppb)', 'NO2 (ppb)', 'NOX (ppb)', 'O3 (ppb)',
                     'PM10 (ug/m3)', 'PM2.5 (ug/m3)', 'PRS (mmHg)', 'RAINF (mm/h)',
                     'RH (%)', 'SO2 (ppb)', 'SR (kW/m2)', 'TOUT (¬∫C)',
                     'WSR (km/h)', 'WDR (azimutal)']

variables_existentes = [col for col in variables_imputar if col in df.columns]

if 'Fecha y hora' in df.columns:
    df['Fecha y hora'] = pd.to_datetime(df['Fecha y hora'])

# Get unique IDs
unique_ids = df['ID'].dropna().unique()

# Dictionary to store results for all stations
resultados_por_estacion = {}
comparacion_final = {}

# ==================================================================================
# LOOP POR CADA ID
# ==================================================================================

for station_id in unique_ids:
    print("\n" + "="*70)
    print(f"ESTACI√ìN: {station_id}")
    print("="*70)

    # Filter data for current station
    df_station = df[df['ID'] == station_id].copy().reset_index(drop=True)

    if len(df_station) < 50:
        print(f"‚ö†Ô∏è Datos insuficientes para la estaci√≥n {station_id} (n={len(df_station)})")
        continue

    print(f"‚úì Dataset: {df_station.shape[0]} observaciones")
    print(f"‚úì Variables: {len(variables_existentes)}")

    # ==================================================================================
    # CREAR SET DE EVALUACI√ìN
    # ==================================================================================

    np.random.seed(123)
    df_eval = df_station[['Fecha y hora'] + variables_existentes].copy().reset_index(drop=True)

    indices_eval = {}
    valores_reales = {}

    for col in variables_existentes:
        indices_disp = df_eval[col].notna()
        n_disp = indices_disp.sum()

        if n_disp > 20:  # Ajustado para estaciones con menos datos
            n_ocultar = int(n_disp * 0.20)
            if n_ocultar > 0:
                indices_ocultar = np.random.choice(
                    df_eval[indices_disp].index,
                    size=n_ocultar,
                    replace=False
                )
                valores_reales[col] = df_eval.loc[indices_ocultar, col].copy()
                df_eval.loc[indices_ocultar, col] = np.nan
                indices_eval[col] = indices_ocultar

    if len(indices_eval) == 0:
        print(f"‚ö†Ô∏è No hay suficientes datos para evaluar en la estaci√≥n {station_id}")
        continue

    print(f"‚úì Variables con evaluaci√≥n: {len(indices_eval)}")

    # ==================================================================================
    # M√âTODO 1: INTERPOLACI√ìN LINEAL
    # ==================================================================================
    print("\n" + "-"*70)
    print("M√âTODO 1: INTERPOLACI√ìN LINEAL")
    print("-"*70)

    df_interp = df_eval.copy()

    for col in variables_existentes:
        df_interp[col] = df_interp[col].interpolate(method='linear', limit_direction='both')
        df_interp[col].fillna(df_interp[col].median(), inplace=True)

    metricas_interp = {}
    for col in indices_eval.keys():
        indices = indices_eval[col]
        y_true = valores_reales[col].values
        y_pred = df_interp.loc[indices, col].values

        metricas_interp[col] = {
            'MAE': mean_absolute_error(y_true, y_pred),
            'RMSE': np.sqrt(mean_squared_error(y_true, y_pred)),
            'R2': r2_score(y_true, y_pred),
            'MAPE': np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100
        }

    df_metricas_interp = pd.DataFrame(metricas_interp).T
    print("\n‚úì Interpolaci√≥n completada")
    print(df_metricas_interp.round(3))

    # ==================================================================================
    # M√âTODO 2: ARIMA
    # ==================================================================================
    print("\n" + "-"*70)
    print("M√âTODO 2: ARIMA")
    print("-"*70)

    df_arima = df_eval.copy()
    metricas_arima = {}

    for col in variables_existentes:
        if col in indices_eval:
            print(f"  Procesando {col}...", end=" ")

            try:
                serie = df_arima[col].copy()
                serie_prep = serie.interpolate(method='linear', limit_direction='both')
                serie_prep.fillna(serie_prep.median(), inplace=True)

                if serie_prep.std() > 0:
                    try:
                        model = ARIMA(serie_prep, order=(2, 1, 2))
                        model_fit = model.fit()
                        predictions = model_fit.fittedvalues
                    except:
                        model = ARIMA(serie_prep, order=(1, 0, 1))
                        model_fit = model.fit()
                        predictions = model_fit.fittedvalues

                    mask_null = serie.isnull()
                    df_arima.loc[mask_null, col] = predictions[mask_null].values

                    indices = indices_eval[col]
                    y_true = valores_reales[col].values
                    y_pred = df_arima.loc[indices, col].values

                    metricas_arima[col] = {
                        'MAE': mean_absolute_error(y_true, y_pred),
                        'RMSE': np.sqrt(mean_squared_error(y_true, y_pred)),
                        'R2': r2_score(y_true, y_pred),
                        'MAPE': np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100
                    }
                    print("‚úì")
                else:
                    print("‚ö†Ô∏è Sin varianza")
                    df_arima[col].fillna(df_arima[col].median(), inplace=True)

            except Exception as e:
                print(f"‚úó")
                df_arima[col].fillna(df_arima[col].median(), inplace=True)

    df_metricas_arima = pd.DataFrame(metricas_arima).T
    print("\n‚úì ARIMA completado")
    if len(df_metricas_arima) > 0:
        print(df_metricas_arima.round(3))

    # ==================================================================================
    # M√âTODO 3: SARIMA
    # ==================================================================================
    print("\n" + "-"*70)
    print("M√âTODO 3: SARIMA (Seasonal ARIMA)")
    print("-"*70)

    df_sarima = df_eval.copy()
    metricas_sarima = {}

    for col in variables_existentes:
        if col in indices_eval:
            print(f"  Procesando {col}...", end=" ")

            try:
                serie = df_sarima[col].copy()
                serie_prep = serie.interpolate(method='linear', limit_direction='both')
                serie_prep.fillna(serie_prep.median(), inplace=True)

                if serie_prep.std() > 0 and len(serie_prep) >= 48:
                    try:
                        model = SARIMAX(
                            serie_prep,
                            order=(1, 1, 1),
                            seasonal_order=(1, 0, 1, 24),
                            enforce_stationarity=False,
                            enforce_invertibility=False
                        )
                        model_fit = model.fit(disp=False, maxiter=50)
                        predictions = model_fit.fittedvalues

                        mask_null = serie.isnull()
                        df_sarima.loc[mask_null, col] = predictions[mask_null].values

                        indices = indices_eval[col]
                        y_true = valores_reales[col].values
                        y_pred = df_sarima.loc[indices, col].values

                        metricas_sarima[col] = {
                            'MAE': mean_absolute_error(y_true, y_pred),
                            'RMSE': np.sqrt(mean_squared_error(y_true, y_pred)),
                            'R2': r2_score(y_true, y_pred),
                            'MAPE': np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100
                        }
                        print("‚úì")

                    except:
                        try:
                            model = SARIMAX(
                                serie_prep,
                                order=(1, 0, 1),
                                seasonal_order=(1, 0, 0, 24),
                                enforce_stationarity=False,
                                enforce_invertibility=False
                            )
                            model_fit = model.fit(disp=False, maxiter=30)
                            predictions = model_fit.fittedvalues

                            mask_null = serie.isnull()
                            df_sarima.loc[mask_null, col] = predictions[mask_null].values

                            indices = indices_eval[col]
                            y_true = valores_reales[col].values
                            y_pred = df_sarima.loc[indices, col].values

                            metricas_sarima[col] = {
                                'MAE': mean_absolute_error(y_true, y_pred),
                                'RMSE': np.sqrt(mean_squared_error(y_true, y_pred)),
                                'R2': r2_score(y_true, y_pred),
                                'MAPE': np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100
                            }
                            print("‚úì (simple)")
                        except:
                            print("‚úó Convergencia")
                            df_sarima[col].fillna(df_sarima[col].median(), inplace=True)
                else:
                    print("‚ö†Ô∏è Datos insuficientes")
                    df_sarima[col].fillna(df_sarima[col].median(), inplace=True)

            except Exception as e:
                print(f"‚úó")
                df_sarima[col].fillna(df_sarima[col].median(), inplace=True)

    df_metricas_sarima = pd.DataFrame(metricas_sarima).T
    print("\n‚úì SARIMA completado")
    if len(df_metricas_sarima) > 0:
        print(df_metricas_sarima.round(3))

    # ==================================================================================
    # M√âTODO 4: HOLT-WINTERS
    # ==================================================================================
    print("\n" + "-"*70)
    print("M√âTODO 4: HOLT-WINTERS (Suavizado Exponencial Triple)")
    print("-"*70)

    df_hw = df_eval.copy()
    metricas_hw = {}

    for col in variables_existentes:
        if col in indices_eval:
            print(f"  Procesando {col}...", end=" ")

            try:
                serie = df_hw[col].copy()
                serie_prep = serie.interpolate(method='linear', limit_direction='both')
                serie_prep.fillna(serie_prep.median(), inplace=True)

                if len(serie_prep) >= 48 and serie_prep.std() > 0:
                    try:
                        model = ExponentialSmoothing(
                            serie_prep,
                            seasonal_periods=24,
                            trend='add',
                            seasonal='add',
                            damped_trend=True
                        )
                        model_fit = model.fit()
                        predictions = model_fit.fittedvalues

                        mask_null = serie.isnull()
                        df_hw.loc[mask_null, col] = predictions[mask_null].values

                        indices = indices_eval[col]
                        y_true = valores_reales[col].values
                        y_pred = df_hw.loc[indices, col].values

                        metricas_hw[col] = {
                            'MAE': mean_absolute_error(y_true, y_pred),
                            'RMSE': np.sqrt(mean_squared_error(y_true, y_pred)),
                            'R2': r2_score(y_true, y_pred),
                            'MAPE': np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100
                        }
                        print("‚úì")
                    except:
                        model = ExponentialSmoothing(serie_prep, trend='add')
                        model_fit = model.fit()
                        predictions = model_fit.fittedvalues

                        mask_null = serie.isnull()
                        df_hw.loc[mask_null, col] = predictions[mask_null].values

                        indices = indices_eval[col]
                        y_true = valores_reales[col].values
                        y_pred = df_hw.loc[indices, col].values

                        metricas_hw[col] = {
                            'MAE': mean_absolute_error(y_true, y_pred),
                            'RMSE': np.sqrt(mean_squared_error(y_true, y_pred)),
                            'R2': r2_score(y_true, y_pred),
                            'MAPE': np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100
                        }
                        print("‚úì (sin estac.)")
                else:
                    print("‚ö†Ô∏è Insuficientes datos")
                    df_hw[col].fillna(df_hw[col].median(), inplace=True)

            except Exception as e:
                print(f"‚úó")
                df_hw[col].fillna(df_hw[col].median(), inplace=True)

    df_metricas_hw = pd.DataFrame(metricas_hw).T
    print("\n‚úì Holt-Winters completado")
    if len(df_metricas_hw) > 0:
        print(df_metricas_hw.round(3))

    # ==================================================================================
    # COMPARACI√ìN DE M√âTODOS PARA ESTA ESTACI√ìN
    # ==================================================================================
    print("\n" + "-"*70)
    print(f"COMPARACI√ìN DE M√âTODOS - ESTACI√ìN {station_id}")
    print("-"*70)

    comparacion = pd.DataFrame({
        'INTERPOLACI√ìN': df_metricas_interp.mean(),
        'ARIMA': df_metricas_arima.mean() if len(df_metricas_arima) > 0 else [np.nan]*4,
        'SARIMA': df_metricas_sarima.mean() if len(df_metricas_sarima) > 0 else [np.nan]*4,
        'HOLT-WINTERS': df_metricas_hw.mean() if len(df_metricas_hw) > 0 else [np.nan]*4
    })

    print(comparacion.round(3))

    mejor_metodo = comparacion.loc['R2'].idxmax()
    print(f"\nüèÜ MEJOR M√âTODO PARA {station_id}: {mejor_metodo} (R¬≤ = {comparacion.loc['R2', mejor_metodo]:.4f})")

    # Guardar resultados
    resultados_por_estacion[station_id] = {
        'comparacion': comparacion,
        'mejor_metodo': mejor_metodo,
        'metricas_interp': df_metricas_interp,
        'metricas_arima': df_metricas_arima,
        'metricas_sarima': df_metricas_sarima,
        'metricas_hw': df_metricas_hw
    }

    comparacion_final[station_id] = comparacion.loc['R2'].to_dict()

    # ==================================================================================
    # VISUALIZACIONES PARA ESTA ESTACI√ìN
    # ==================================================================================

    # Gr√°fico 1: Comparaci√≥n de m√©tricas
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    fig.suptitle(f'Comparaci√≥n de M√©todos - Estaci√≥n {station_id}',
                 fontsize=16, fontweight='bold')

    metricas_plot = ['MAE', 'RMSE', 'R2', 'MAPE']
    for idx, metrica in enumerate(metricas_plot):
        ax = axes[idx//2, idx%2]

        data_plot = pd.DataFrame({
            'INTERPOLACI√ìN': df_metricas_interp[metrica] if metrica in df_metricas_interp.columns else np.nan,
            'ARIMA': df_metricas_arima[metrica] if len(df_metricas_arima) > 0 and metrica in df_metricas_arima.columns else np.nan,
            'SARIMA': df_metricas_sarima[metrica] if len(df_metricas_sarima) > 0 and metrica in df_metricas_sarima.columns else np.nan,
            'HOLT-WINTERS': df_metricas_hw[metrica] if len(df_metricas_hw) > 0 and metrica in df_metricas_hw.columns else np.nan
        })

        data_plot.plot(kind='bar', ax=ax, rot=45, width=0.8)
        ax.set_title(f'{metrica} por Variable', fontweight='bold', fontsize=12)
        ax.set_ylabel(metrica)
        ax.legend(title='M√©todo', fontsize=8, loc='best')
        ax.grid(axis='y', alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Gr√°fico 2: Ranking
    fig, ax = plt.subplots(figsize=(10, 6))

    ranking = comparacion.loc['R2'].dropna().sort_values(ascending=True)
    colors = ['green' if x == ranking.max() else 'steelblue' for x in ranking]
    ranking.plot(kind='barh', ax=ax, color=colors)
    ax.set_title(f'Ranking por R¬≤ - Estaci√≥n {station_id}', fontsize=14, fontweight='bold')
    ax.set_xlabel('R¬≤ (Coeficiente de Determinaci√≥n)')
    ax.axvline(x=0.7, color='red', linestyle='--', label='Objetivo (0.7)')
    ax.legend()
    ax.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()

# ==================================================================================
# RESUMEN FINAL PARA TODAS LAS ESTACIONES
# ==================================================================================
print("\n" + "="*70)
print("RESUMEN FINAL - TODAS LAS ESTACIONES")
print("="*70)

if len(comparacion_final) > 0:
    df_comparacion_final = pd.DataFrame(comparacion_final).T
    print("\nR¬≤ por Estaci√≥n y M√©todo:")
    print(df_comparacion_final.round(3))

    print("\n\nPromedio por M√©todo (todas las estaciones):")
    print(df_comparacion_final.mean().round(3))

    mejor_metodo_global = df_comparacion_final.mean().idxmax()
    print(f"\nüèÜ MEJOR M√âTODO GLOBAL: {mejor_metodo_global} (R¬≤ promedio = {df_comparacion_final.mean()[mejor_metodo_global]:.4f})")

    # Gr√°fico final: Heatmap R¬≤ por estaci√≥n
    fig, ax = plt.subplots(figsize=(12, max(6, len(unique_ids)*0.5)))
    sns.heatmap(df_comparacion_final, annot=True, fmt='.3f', cmap='RdYlGn',
                center=0.5, vmin=0, vmax=1, ax=ax, cbar_kws={'label': 'R¬≤'})
    ax.set_title('R¬≤ por Estaci√≥n y M√©todo de Imputaci√≥n',
                 fontsize=14, fontweight='bold')
    ax.set_xlabel('M√©todo')
    ax.set_ylabel('Estaci√≥n')
    plt.tight_layout()
    plt.show()

    # Gr√°fico final: Comparaci√≥n promedio
    fig, ax = plt.subplots(figsize=(10, 6))
    df_comparacion_final.mean().sort_values(ascending=True).plot(kind='barh', ax=ax, color='steelblue')
    ax.set_title('R¬≤ Promedio por M√©todo (Todas las Estaciones)', fontsize=14, fontweight='bold')
    ax.set_xlabel('R¬≤ Promedio')
    ax.axvline(x=0.7, color='red', linestyle='--', label='Objetivo (0.7)')
    ax.legend()
    ax.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()

    # Tabla resumen de mejor m√©todo por estaci√≥n
    print("\n\nMejor M√©todo por Estaci√≥n:")
    for station_id, resultados in resultados_por_estacion.items():
        metodo = resultados['mejor_metodo']
        r2 = resultados['comparacion'].loc['R2', metodo]
        print(f"  Estaci√≥n {station_id}: {metodo} (R¬≤ = {r2:.4f})")

print("\n‚úÖ An√°lisis completo para todas las estaciones")
```
